```python
from selenium.webdriver.chrome.service import Service as ChromeService
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
from random import randint
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
import bs4

options = webdriver.ChromeOptions()

# Uncomment this line if you want to run Chrome in headless mode
# options.add_argument("--headless")

pagination_url = 'https://uttoron.academy/question/45th-bcs/'

start = time.time()

# Initialize the ChromeDriver using ChromeDriverManager
driver = webdriver.Chrome(service=ChromeService(ChromeDriverManager().install()), options=options)

driver.get(pagination_url)

# Wait for the page to load completely
WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CLASS_NAME, "question-item")))

html = driver.page_source
soup = bs4.BeautifulSoup(html, 'html.parser')

output = []

question_div = soup.find_all("div", {"class": "question-item"})
for question_sections in question_div:
    question = question_sections.find("div", {"class": "question"}).find("p")
    question = question.text.lstrip().rstrip().replace("\n", "")
    print(question)
    count = 0

    answers = question_sections.find("div", {"class": "options"}).find_all("li")
    for answer in answers:
        answer = answer.find_all("span")[1]
        answer = answer.text.lstrip().rstrip().replace("\n", "")
        print(answer)
        count += 1
        if count == 1:
            answer1 = answer

        if count == 2:
            answer2 = answer

        if count == 3:
            answer3 = answer

        if count == 4:
            answer4 = answer

    right_answer = question_sections.find("li", {"class": "is_right"})
    if right_answer:
        right_answer = right_answer.find_all("span")[1]
        right_answer = right_answer.text.lstrip().rstrip().replace("\n", "") if right_answer else ''
        print("---->", right_answer)
    else:
        right_answer = ""

    output.append({
        'question': question,
        'answer 1': answer1,
        'answer 2': answer2,
        'answer 3': answer3,
        'answer 4': answer4,
        'correct_answer': right_answer
    })

print(output)
driver.quit()
```
