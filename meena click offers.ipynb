```python
import requests
import bs4
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.common.keys import Keys
from webdriver_manager.chrome import ChromeDriverManager
import time
import math
import re
url = 'https://meenaclick.com'
driver = webdriver.Chrome(ChromeDriverManager().install())
driver.get(url)
driver.maximize_window()
output = []
wait = WebDriverWait(driver, 30)
wait.until(EC.visibility_of_element_located((By.CLASS_NAME, "ant-menu")))
html = driver.page_source
soup = bs4.BeautifulSoup(html)

for category in soup.find("div", {"class": "home-menu"}).find_all("li", {"class": "ant-menu-item"}):
    print(category.text)
    link = category['routerlink']
#     print(link)
    x = driver.find_element_by_css_selector("li[routerlink='{}']".format(link))
#     print(x)
    x.click()
    time.sleep(5)
    i = 1
    while True:
        current_page_number = i
        print(f"Processing page {current_page_number}..")

        try:
            next_page_link = driver.find_element_by_class_name("ng-star-inserted").find_element_by_link_text(str(i))
            next_page_link.click()
            i = i+1
    #         time.sleep(5)
            wait = WebDriverWait(driver, 30)
            wait.until(EC.visibility_of_element_located((By.CLASS_NAME, "img")))
            html = driver.find_element_by_tag_name('html')
            html.send_keys(Keys.HOME)
            time.sleep(10)
            check_height = driver.execute_script("return document.body.scrollHeight;") 
            while True:
            #     driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
                html.send_keys(Keys.END)
                time.sleep(10)
                height = driver.execute_script("return document.body.scrollHeight;") 
                if height == check_height: 
                    break 
                check_height = height
            html = driver.page_source
            soup = bs4.BeautifulSoup(html)
            # driver.close()

            for product in soup.find_all("div", {"class": "main-img"}):
                image = product.find("img")
                title = product.find("h1",{"class": "title"})
                price = product.find_all("span",{"class": "reg-price"})
                discount = product.find_all("span",{"class": "off-price"})
                link = product.find("a",{"class": "main-link"})
                image = image['src']
                productName = title.text
                pricetxt = price[0].text if len(price) else '0'
                discounttxt = discount[0].text if len(discount) else '0'
                price = re.findall(r"[-+]?\d*\.\d+|\d+", pricetxt)
                discount = re.findall(r"[-+]?\d*\.\d+|\d+", discounttxt)
                discount_percent = math.ceil(float(discount[0])/float(price[0]))*100 if discounttxt!='0' else 0
                link = product.find_parent('div').find('a',{"class": "main-link"})['href']
                output.append({
                    'name' : productName,
                    'image': image,
                    'quantity' : '',
                    'price': price,
                    'discount': discount,
                    'link': link,
                    'category': category.text
                })
#                 print(image)

        except NoSuchElementException:
            print(f"Exiting. Last page: {current_page_number}.")
            break

    driver.find_element_by_css_selector("a[routerlink='/']").click()
    time.sleep(2)
    
driver.close()
# print(soup.prettify())
```
