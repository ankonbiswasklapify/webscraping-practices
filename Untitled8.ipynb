```python
from selenium import webdriver
import pandas as pd
from selenium.webdriver.common.by import By
import time
from webdriver_manager.chrome import ChromeDriverManager
url="https://www.youtube.com/c/LearnwithSumit"
driver = webdriver.Chrome(ChromeDriverManager().install())
driver.get(url)
driver.maximize_window()
video_list=[]
videos=driver.find_elements(By.CLASS_NAME,"style-scope ytd-grid-video-renderer")
for video in videos:
    tittle=video.find_element(By.ID,"video-title").text
    time.sleep(3)
    channel_name=video.find_element(By.ID,"text-container").text
    views=video.find_element(By.ID,"metadata-line").text
    print(tittle,channel_name,views)
    video_item={
        'tittle':'tittle',
        'channel_name':'channel_name',
        'views':'views'
    }
    video_list.append(video_item)
    
df=pd.DataFrame(video_list)    
print(df)  
```
