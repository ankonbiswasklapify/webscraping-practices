```python
import requests
import bs4
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from webdriver_manager.chrome import ChromeDriverManager
output = []
url = 'https://www.sattacademy.com/job-solution/view?cat_id=3&sub_cat_id=848'
driver = webdriver.Chrome(ChromeDriverManager().install())
driver.get(url)
wait = WebDriverWait(driver, 30)
wait.until(EC.visibility_of_element_located((By.CLASS_NAME, "card-title")))
html = driver.page_source
soup = bs4.BeautifulSoup(html)
# print(soup)
i = 0
for products in soup.find_all("div", {"class": "col-sm-12"}):
#     i= i+1
#     print(i)
    obj = {}
    question_text = products.find("a", {"class": "nav-link"})
    if(question_text):
        question_text = question_text.text
        print(question_text)
    answers_divs = products.find_all("p", {"class": "answer"})
    answers_arr = []
    for answers in answers_divs:
        print(answers.text)
        answers_arr.append(answers.text)
    right_answer = products.find("p", {"class": "answer rightAns"})
    if(right_answer):
        right_answer = right_answer.text
        print('right->', right_answer)
    obj = {
        'question': question_text,
        'answers': answers_arr,
        'right_answer': right_answer
    }
    output.append(obj)

print(output)
```
