```python

from select import select
from selenium.webdriver.support.ui import Select
from selenium.webdriver.common.keys import Keys
import bs4
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from webdriver_manager.chrome import ChromeDriverManager
url='https://www.meenaclick.com/'
driver = webdriver.Chrome(ChromeDriverManager().install())
driver.get(url)
driver.maximize_window()
time.sleep(5)
html=driver.page_source
soup=bs4.BeautifulSoup(html)
for category in soup.find("div", {"class": "home-menu"}).find_all("li", {"class": "ant-menu-item"}):
    print(category.text)    
    # print(category.text)
    link = category['routerlink']
    x = driver.find_element(By.CSS_SELECTOR,"li[routerlink='{}']".format(link)) 
    x.click()
    time.sleep(5)
    # driver.back()
    # time.sleep(10)

    html1=driver.page_source
    soup1=bs4.BeautifulSoup(html1)
    abc=soup1.find_all("ul",{"class":"list"})   #list
    if len(abc):
        sub_category=abc[0].find_all("li",{"class":"item ng-star-inserted"})  #item ng-star-inserted
        # print(sub_category)
        time.sleep(5)
        # driver.back()
        for link in sub_category:
            cate=link.find("a").text.rstrip()
            print(cate)
            wait = WebDriverWait(driver,30)
            wait.until(EC.element_to_be_clickable((By.XPATH,"//div[@class='box ng-star-inserted']//a[@class='link'][normalize-space()='"+cate+"']"))).click()
            time.sleep(10)
            top = driver.execute_script("return document.body.scrollHeight;") 
            while True:
                driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
                time.sleep(2)
                height = driver.execute_script("return document.body.scrollHeight;") 
                if height == top: 
                    break 
                top = height 
                
            driver.back()
            time.sleep(10)
    driver.back()
    driver.implicitly_wait(10)
    time.sleep(5)
    # print('-----')

    

    

     
   
    



```
